## c) greedy, optimistic initial values, incremental simple average

```python
evaluator = p1.Evaluator(p1.EpsilonGreedyAgent, epsilon=0.0, iterations=5000, update_method='incremental', action_values=np.ones(10)*10)
cp, cr = evaluator.run()
evaluator.visualize()
```

## Greedy and eGreedy comparison

```python
fig = plt.figure()
plt.grid(zorder=0)
plt.xlabel("Time step")
plt.ylabel("Optimal action percentage")
plt.plot(ap, label= f"epsilon=0.00 - max={ap.max():.2f}")
plt.plot(b1p, label=f"epsilon=0.01 - max={b1p.max():.2f}")
plt.plot(b2p, label=f"epsilon=0.05 - max={b2p.max():.2f}")
plt.plot(b3p, label=f"epsilon=0.10 - max={b3p.max():.2f}")
plt.plot(b4p, label=f"epsilon=0.15 - max={b4p.max():.2f}")
plt.plot(b5p, label=f"epsilon=0.20 - max={b5p.max():.2f}")
plt.plot(cp, label= f"e=0 optimistic-max={cp.max():.2f}")
plt.legend(bbox_to_anchor=(1.52, 1), loc='upper right')
plt.show();
```
